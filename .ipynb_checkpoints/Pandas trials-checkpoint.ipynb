```python
def splitname(row):
    row['First'] = row['President'].split(" ")[0]
    row['Last'] = row['President'].split(" ")[-1]
    return row

df = df.apply(splitname, axis = "columns")
```

```python
df.head()
```

```python
del(df['First'])
del(df['Last'])
```

```python
df.head()
```

```python
pattern = "(?P<First>[\w]*)(?:.* )(?P<Last>[\w]*$)"
```

```python
df['President'].str.extract(pattern).head() 
```

```python
df['First'] = df['President'].str.extract(pattern)["First"]
df['Last'] = df['President'].str.extract(pattern)["Last"]
```

```python
df.head()
```

```python
df['Born'] = df['Born'].str.extract("([\w]{3} [0-9]{1,2}, [0-9]{4})")
```

```python
df["Born"] = pd.to_datetime(df['Born'])
```

```python
df1 = pd.DataFrame([np.random.random(100) for x in range(100)])
df1.head()
```

```python
df2 = pd.DataFrame([np.random.random(100) for x in range(100)])
df2.head()
```

```python
def test_columns(alpha = 0.1):
    num_diff = 0
    
    for col in df1.columns:
        teststat,pval = ttest_ind(df1[col], df2[col])
        
        if pval <= alpha:
            print("Columns {} is statistically significantly different at alpha = {} and pval = {}".format(col, alpha, pval))
            num_diff += 1
    print('Total number of different columns is {} which is {}%'.format(num_diff, num_diff/len(df1.columns)*100))
```

```python
test_columns(alpha = 0.05)
```

```python
df2 = pd.DataFrame([np.random.chisquare(df= 1, size = 100) for x in range(100)])
```

```python
test_columns(alpha = 0.05)
```
